### Calculate 95 % Confidence Interval using Z-statistic

```python
# First I will build a sampler to sample the values of the charges column and calculate means for each sample
def sampler(data, n):
    return np.random.choice(data, n)
```

```python
s=pd.DataFrame({'charges':[sampler(medical.charges, 1338).mean() for i in range(1000)]})
```

```python
s.info()
```

```python
s.head()
```

```python
# Calculate the mean of the sample
mean_charges=np.mean(s.charges)
mean_charges
```

```python
# Estimate the standard deviation of the population (r_mean=r_population/sqrt(n))
np.std(s.charges, ddof=1)*np.sqrt(1300)
```

```python
# Calculate the critical value for 95% confidence interval
# 95 % confidence level is going to be used to find the critical value and first we need to calculate a (alpha)
a=1-0.95 
area_one_tail=a/2 
area_outside_the_tail = 1-a/2
area_outside_the_tail
```

```python
# Calculate the critical value from z-table
from scipy.stats import norm
cv=norm.ppf(0.975)
cv
```

```python
# Calculate the Standard Error which is calculated as sample_std/sqrt(sample_size)
se=np.std(s.charges, ddof=1)/np.sqrt(len(s))
se
```

```python
# Calculate the Margin of Error
me=cv*se
me
```

```python
CI=mean_charges-me, mean_charges+me
CI
```

### Calculate the 95% Confidence Interval with Bootsrapping 

```python
# Sample the original data set with replacment for 1000 times
bootstrap=pd.DataFrame({'charges':[medical.charges.sample(1338, replace=True).mean() for i in range(1000)]})
```

```python
# Look at the histogram of the sampled means with the mean from the original data sketched on top in red color
bootstrap.hist(histtype='step')
plt.axvline(np.mean(medical.charges), color='r', linestyle='--')
```

```python
# Calculate 95% confidence interval for the true mean
bootstrap.charges.quantile(0.025), bootstrap.charges.quantile(0.975)
```

Bootstrapping gave as a wider interval than the one we calculated from using Z-statistic. The average charge below 12,000 is observed in none of the confidence intervals calculated above therefore there the administrator of the hospital should not be concerned about revenue falling under the 12,000

__Q4:__ State the null and alternative hypothesis here. Use the _t_-test for the difference between means, where the pooled standard deviation of the two groups is given by:
\begin{equation}
s_p = \sqrt{\frac{(n_0 - 1)s^2_0 + (n_1 - 1)s^2_1}{n_0 + n_1 - 2}}
\end{equation}

and the *t*-test statistic is then given by:

\begin{equation}
t = \frac{\bar{x}_0 - \bar{x}_1}{s_p \sqrt{1/n_0 + 1/n_1}}.
\end{equation}

(If you need some reminding of the general definition of ***t-statistic***, check out the definition on p. 404 of *AoS*). 

What assumption about the variances of the two groups are we making here?

__A:__ We assume they have similar variances

Hypothesis test:

H_0: There is no statistically significant difference between what is charged to people with insurance than those without

H_1: There is statistically significant difference between what is charged to people with insurance than those without

#### Step 1: Prepare the data and draw samples

```python
medical.head()
```

```python
with_insurance=medical[medical['insuranceclaim']==1]
without_insurance=medical[medical['insuranceclaim']==0]
```

```python
with_insurance.charges.mean()
```

```python
without_insurance.charges.mean()
```

```python
len(with_insurance), len(without_insurance)
```

```python
s_without_insurance=sampler(without_insurance.charges,31)
s_with_insurance=sampler(with_insurance.charges, 31)
```

#### Step 2: Calculate the t-value

```python
n0=len(s_without_insurance)
n1=len(s_with_insurance)
s0=np.std(s_without_insurance)
var_0=s0**2
s1=np.std(s_with_insurance)
var_1=s1**2
df=n0+n1-2
sp=np.sqrt(((n0-1)*(var_0)+(n1-1)*(var_1))/(df))
sp
```

```python
x0=np.mean(s_without_insurance)
x1=np.mean(s_with_insurance)
t=(x0-x1)/(sp*np.sqrt((1/n0)+(1/n1)))
t
```

```python
from scipy import stats
t, p = stats.ttest_ind(s_without_insurance, s_with_insurance, equal_var=True)
```

```python
t, p
```

t value of -2.81 is smaller than 2.0 therefore we fail to reject the Null hypothesis and therefore would conclude that there is no statistically significant difference between what is charged to people with insurance than those without
